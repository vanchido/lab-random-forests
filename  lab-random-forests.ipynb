```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report

import warnings
warnings.filterwarnings('ignore')
```

```python
target = pd.read_csv('files_for_lab/target.csv')
cat = pd.read_csv('files_for_lab/categorical.csv')
num = pd.read_csv('files_for_lab/numerical.csv')
```

```python
cat = pd.get_dummies(cat, columns=cat.columns, drop_first=True)
```

```python
# Join the cat & num dataframes
X = pd.concat([num, cat], axis=1)
```

```python
y_regression = target['TARGET_D']
y_classification = target['TARGET_B']
```

```python
# Train-test split for the classification model
X_train, X_test, y_ctrain, y_ctest = train_test_split(X, y_classification, test_size=0.3, random_state=42)
```

```python
# Oversampling using SMOTE
X_sm, y_csm = SMOTE().fit_resample(X_train, y_ctrain)
```

```python
# Apply the Random Forests algorithm
model = RandomForestClassifier(max_depth=10, n_estimators=100, random_state=42)
model.fit(X_sm, y_csm)
```

```python
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]
```

```python
model.score(X_test, y_ctest)
```

```python
print(classification_report(y_ctest, y_pred))
```

```python
# Train-test split for the regression model
X_train, X_test, y_rtrain, y_rtest = train_test_split(X, y_regression, test_size=0.3, random_state=42)
```

```python
# Oversampling using SMOTE
X_sm, y_rsm = SMOTE().fit_resample(X_train, y_ctrain)
```

```python
# Apply the Random Forests algorithm
model = RandomForestClassifier(max_depth=10, n_estimators=100, random_state=42)
model.fit(X_sm, y_rsm)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_ctest, y_pred))
```
